This package use for Image processing "cv2" is like "opencv" package

```python
import numpy as np
import cv2  as cv
import matplotlib.pyplot as plt
```

"imread" uses for reading image
"plt.imshow" uses for showing image

```python
I1=cv.imread('Image/Car1.jpg')
plt.imshow(I1)
```

If we want to know about of amount of each pixel use [x,y,(RGB)]

```python
print('RGB for Pixel (1,1)=',I1[1,1,:])
print('The number of x and y pixels=',I1.shape)
print('Kind of image=',I1.dtype)
```

This command convert RGB Image to Grayscale Image

```python
I2=cv.cvtColor(I1, cv.COLOR_BGR2GRAY) 
plt.imshow(I2)
```

```python
print('RGB for Pixel (1,1)=',I2[1,1])
print('The number of x and y pixels=',I2.shape)
print('Kind of image=',I2.dtype)
```

"cv.imwrite" make Image

```python
cv.imwrite('Image/Car1_gray.jpg',I2)
```

"np.zeros(x,y)" uses for making a zero array

```python
S=I1.shape
#Red
I1_R=np.zeros((S[0],S[1],3),dtype=np.uint8)
I1_R[:,:,0]=I1[:,:,0]
plt.subplot(1,3,1)
plt.imshow(I1_R)
plt.axis('off')
plt.title('Red')
#Green
I1_G=np.zeros((S[0],S[1],3),dtype=np.uint8)
I1_G[:,:,1]=I1[:,:,1]
plt.subplot(1,3,2)
plt.imshow(I1_G)
plt.axis('off')
plt.title('Green')
#Blue
I1_B=np.zeros((S[0],S[1],3),dtype=np.uint8)
I1_B[:,:,2]=I1[:,:,2]
plt.subplot(1,3,3)
plt.imshow(I1_B)
plt.axis('off')
plt.title('Blue')
```

Convert grayscale image to binary image by using loop
I2 is a grayscale image which is uint8 image (each pixel is between 0 to 255)

```python
Th=120 #threshold for converting grayscale image to binary image
S=I2.shape
I3=np.zeros((S[0],S[1]),dtype=int)
for i1 in range(S[0]):
    for i2 in range(S[1]):
        if I2[i1,i2]>Th:
            I3[i1,i2]=1
plt.imshow(I3)
plt.axis('off')
plt.title('Binary image with threshold='+str(Th))
```

Convert grascale image to binary image with special code

```python
Th=120
ret,I4=cv.threshold(I2,Th,255,0)
plt.imshow(I4)
plt.axis('off')
plt.title('Binary of image with "cv.threshold" code and threshold is ('+str(Th)+')')
```

This part separate a special part of image by two method which shows below

```python
A=[[70, 100],[100, 180]]
#Method 1
I5=np.zeros((A[0][1]-A[0][0],A[1][1]-A[1][0]),dtype=np.uint8)
x=int(-1)
for i1 in range(A[0][0],A[0][1],1):
    x=x+1
    y=int(-1)
    for i2 in range(A[1][0],A[1][1],1):
        y=y+1
        I5[x,y]=I2[i1,i2]
plt.subplot(1,2,1)
plt.imshow(I5)
plt.axis('off')
#Method 2
plt.subplot(1,2,2)
plt.imshow(I2[A[0][0]:A[0][1],A[1][0]:A[1][1]])
plt.axis('off')
```
